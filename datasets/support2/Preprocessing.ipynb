## Preprocessing
- One-hot encode variable: sex(2), dzclass(4), race (5), ca(3), income(4), sfdm2 (5)
- Target variables: hospdead(0 / 1)
- Remove features: death (non-target), slos, d.time(related directly to outcome), dzgroup (similar to dzclass), charges, totcst, totmcst (cost not related), not to use previous model(aps, sps, surv2m, surv6m) and prg2m, prg6m, dnr, dnrday.

Total features: 47 + 17 - 16 = 48

```python
one_hot_encode_cols = ['sex', 'dzclass', 'race' , 'ca', 'income', 'sfdm2']
target_variables = ['hospdead']
remove_features = ['death', 'slos', 'd.time', 'dzgroup', 'charges', 'totcst', 
                   'totmcst', 'aps', 'sps', 'surv2m', 'surv6m', 'prg2m', 'prg6m', 'dnr', 'dnrday']
```

```python
df = df.drop(remove_features, axis=1)
df.shape
```

```python
rest_colmns = [c for c in df.columns if c not in (one_hot_encode_cols + target_variables)]
df[rest_colmns].describe()
```

```python
# Normalize to 0 mean 1 std
df[rest_colmns] = df[rest_colmns].apply(lambda x: (x - x.mean()) / x.std())
# Cap the value to +-3
df.loc[:, rest_colmns][df.loc[:, rest_colmns] > 3.] = 3.
df.loc[:, rest_colmns][df.loc[:, rest_colmns] < -3.] = -3.
# Impute the missing values for 0.
df[rest_colmns] = df[rest_colmns].fillna(0.)
```

```python
# Handle binary attributes
df = pd.get_dummies(df, prefix=one_hot_encode_cols)
```

```python
x = df.drop(['hospdead'], axis=1)
y = df[['hospdead']]
```

```python
x.to_csv('x.csv', index=None)
y.to_csv('y.csv', index=None)
```
